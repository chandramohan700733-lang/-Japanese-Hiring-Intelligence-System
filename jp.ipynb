```python
import numpy as np
```

```python
#japanese city preference
japan_cities = [
    "Tokyo", "Osaka", "Kyoto",
    "Yokohama", "Nagoya",
    "Fukuoka", "Sapporo"
]
```

```python
VISA_Status = ["Approved", "Pending", "Denied"]
JLPT_levels = ["N1", "N2", "N3", "N4", "N5"]
RELOCATION_Ready = ["Yes", "No"]
```

```python
df ["Prefered_Japan_City"] = np.random.choice(japan_cities, size=len(df))
```

```python
df["JLPT_Level"] = np.random.choice(JLPT_levels, size=len(df), p=[0.40, 0.30, 0.20, 0.10, 0.00])
```

```python
df["VISA_Status"] = np.random.choice(VISA_Status, size = len(df), p=[0.50, 0.30, 0.20])
```

```python
print(df.head())
```

```python
df.rename(columns={"Prefered_Japan_City": "Preferred_Japan_City"}, inplace=True)
```

```python
print(df.head())
```

```python
df["VISA_Status"]= df["VISA_Status"].replace("Denied","Rejected")
```

```python
Interview_Status = ["Selected","Rejected", "Under Review", "Interview Scheduled"]
df["Interview_Status"]=np.random.choice(Interview_Status, size =len(df),p=[0.25,0.30,0.20,0.25])
```

```python
Communication = ["Beginner","Intermediate","Advanced"]
df["Communication"]= np.random.choice(Communication, size = len(df), p = [0.40, 0.40, 0.20])
```

```python

# Skill Category
skills = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Machine Learning",
    "Data Analysis"
]

df["Skill_Category"] = np.random.choice(
    skills,
    size=len(df)
)

# Hiring Score
df["Hiring_Score"] = np.random.randint(
    50,
    100,
    size=len(df)
)

# Preview
print(df.head())
```

```python
df.rename(columns={"RELOCATION_Ready": "Relocation_Ready",
                   "Communication": "Japanese_Communication",
                   "VISA_Status": "Visa_Status",
                   "Interview_Status": "Interview_Status"}, inplace=True)
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
print(df.columns)
```

```python
df["salary_band"] = pd.cut(
    df["Salary_INR"],
    bins=[300000,700000,1200000,1800000,2400000,3000000],
    labels=[
        "Entry Level",
        "Mid Level",
        "Senior Level",
        "Manager Level",
        "Leadership Level"
    ],
    include_lowest=True
)

df["Experience_Group"] = pd.cut(df["Experience_Years"], bins = [-1, 2, 5, 10, 20], labels = ["Fresher", "Junior Level", "Mid Level", "Senior Level"])
```

```python
df["Hire_Date"] = pd.to_datetime(df["Hire_Date"])

df["Hire_Year"] = df["Hire_Date"].dt.year
df["Hire_Month"] = df["Hire_Date"].dt.month_name()
df["Hire_Quarter"] = "Q" + df["Hire_Date"].dt.quarter.astype(str)

print(df[["Hire_Date","Hire_Year","Hire_Month","Hire_Quarter"]].head())
```

```python
df.to_csv(
    "Japanese_Hiring_Intelligence_System.csv",
    index=False
)
```
